```python
def diff_function(data):
        df_cat = pd.DataFrame()
        df_cont = pd.DataFrame()
        df_disc = pd.DataFrame()
        
        for i in data:  
        
            if np.dtype(data[i]) == 'object':
                df_cat[i] = data[i]
        
            elif len(data[i].unique()) > 50 or np.dtype(data[i]) == 'float':  #(len(data[i].unique())*0.8):
                df_cont[i] = data[i]
        
            else: 
                df_disc[i] = data[i]

        return df_cat, df_cont, df_disc
```
